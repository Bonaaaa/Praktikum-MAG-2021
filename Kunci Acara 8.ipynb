# GANJIL

```python
import numpy as np
import matplotlib.pyplot as plt

def naive_DFT (x):
#fungsi DFT
    N = np.size(x)
    X = np.zeros(N,dtype=np.complex128)
    for m in range (0,N):
        for n in range (0,N):
            X[m] += x[n]*np.exp(-np.pi*2j*m*n/N)
    return X

def naive_IDFT (x):
#fungsi IDFT
    N = np.size(x)
    X = np.zeros(N, dtype=np.complex128)
    for m in range (0,N):
        for n in range (0, N):
            X[m] += x[n]*np.exp (-np.pi*2j*m*n/N)
    return X/N

N = 1000; #Jumlah sampel (sampling) data sinus
dt = 1/200 #interval sampling data sinus
t = np.linspace (0, N*dt, N)
#runtun waktu fungsi sinus berada pada rentang 0 sampai hasil kali sampling dengan intervalnya dan interval berupa interval samplingnya dipanggil sebagai t
ysin = np.sin(50*2*np.pi*t) + 0.5*np.sin(80*2*np.pi*t)
#fungsi sinus dipanggil sebagai ysin
ysinf = naive_DFT (ysin)
#DFT fungsi superposisi dipanggil sebagai ysinf
xsinf = np.linspace (0, 1/(2*dt), N//2)
#rentang sumbu x DFT fungsi superposisi dipanggil sebagai xsinf
ysinbalik = naive_IDFT (ysinf)
#IDFT fungsi superposisi dipanggil sebagai ysinbalik

plt.figure(figsize=(10,10))
plt.subplot(3,1,1)
plt.plot(t,ysin,'b')
plt.title('Data Mentah')
plt.ylabel('Amplitudo')
plt.xlabel('Waktu (s)')

plt.subplot(3,1,2)
plt.plot(xsinf, 2/N*np.abs(ysinf[0:N//2]),'r-')
plt.title('Power Spectral')
plt.xlabel('Frekuensi (Hz)')
plt.ylabel('Spectral')
plt.subplot(3,1,3)

plt.plot(t,ysinbalik,'b-')
plt.title('Data Hasil IDFT')
plt.ylabel('Amp')
plt.xlabel('Waktu (s)')
plt.grid()
plt.tight_layout()
plt.show()
```
