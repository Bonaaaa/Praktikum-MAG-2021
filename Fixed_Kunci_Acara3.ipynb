```python
#SOAL 1A
import sympy #memanggil modul sympy

t, s = sympy.symbols('t, s')#memanggil variabel waktu dan frekuensi kompleks sebagai t dan s 
f = (1+sympy.exp(3*t))/(sympy.exp(-3*t)) #memanggil fungsi domain waktu soal sebagai f
f #menampilkan soal 1A
F = sympy.laplace_transform(f,t,s, noconds=True) #memanggil hasil transformasi laplace dari f sebagai F
F #menampilkan transformasi laplace soal 1A
#Fbalik = sympy.inverse_laplace_transform(F,s,t)#melakukan invers transformasi laplace
#Fbalik
```

```python
#SOAL 2A
import sympy #memanggil modul sympy
import scipy.signal as signal # mengimport modul scipy.signal sebagai signal
from scipy.signal import lsim # mengimport lsim dari modul scipy.signal
import pylab as plt # mengimport modul pylab sebagai plt

t, s = sympy.symbols('t, s')#memanggil variabel waktu dan frekuensi kompleks sebagai t dan s 
f2a = (1)+(1/s)+(1/s**2)+(1/s**3)+(1/s**4)+(1/s**5) #memanggil fungsi domain waktu soal sebagai f
f2a #menampilkan soal 2A
```

```python
Fbalik = sympy.inverse_laplace_transform(f2a,s,t)#melakukan invers transformasi laplace soal 2A
Fbalik #menampilkan invers transformasi laplace balik soal 2A
```

```python
dt = 0.01 # interval waktu sinyal adalah 0.01 detik per data
t = plt.arange(0, 5+dt, dt) # didefiniskan interval waktu dari 0 hingga 5 
numerator = [1, 1, 1, 1, 1, 1] # mendefinisikan pembilang sebagai numerator
denumerator = [1, 0, 0, 0, 0, 0] # mendefinisikan penyebut sebagai denumerator
X = signal.TransferFunction(numerator, denumerator) # mendefinisikan fungsi transfer sebagai X
x = signal.impulse(X, T = t) # mendefinisikan impulse respon sebagai x
plt.plot(t, x[1],'b') # melakukan plot
plt.title('Inversi Laplace Transform Signal') # memberi judul plot
plt.grid() # memberikan kisi-kisi pada plot
plt.xlabel('Waktu (s)') # memberi label sumbu x
plt.ylabel('Amplitudo') # memberi label sumbu y
plt.show() # menampilkan plot
```
