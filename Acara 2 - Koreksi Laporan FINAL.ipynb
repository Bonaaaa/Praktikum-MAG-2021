```python
import numpy as np #memanggil modul numpy sebagai np
import pylab as plt #memanggil modul pylab sebagai plt

#Interval waktu
dt = 0.01 #interval waktu sinyal adalah 0.01 detik per data
t  = np.arange(-5,5+dt,dt) #rentang waktu data dari -5 sampai 5+interval waktu sinyal dengan interval waktu 0,01 detik

#sinyal masukan x(t)
x1s1 = t #sinyal masukan ke S1 didefinisikan sebagai "t"

#1. keluaran S1 adalah y1s1 = sin (t) dan y2s1 = t
y1s1 = np.sin(t) #sinyal respon sistem S1 adalah "sin (t)" dipanggil sebagai y1s1
y2s1 = t #sinyal respon sistem S1 adalah "t" dipanggil sebagai y2s1

#plot respon sistem S1
fig, ax = plt.subplots() #plot
plt.plot (x1s1, y1s1, label = "$y_1(t)$") #plot respon sistem S1 untuk y1
plt.plot (x1s1, y2s1, label = "$y_2(t)$") #plot respon sistem S1 untuk y2
plt.title ("1.Sinyal Respon S1 - SIMO") #memberikan judul pada sinyal respon sistem s1
plt.legend() #memberikan legenda pada plot sinyal respon sistem s1
plt.grid(True) #membuat grid pada plot

#2. keluaran S2 adalah y1s2 = y1s1 * y2s1 dan y2s2 = -y1s1 * y2s1 
#sinyal masukan S2
x1s2 = np.sin(t) #sinyal masukan x1 S2 adalah sinyal respons y1s1 "sin (t)" dipanggil sebagai x1s2
x2s2 = t #sinyal masukan x2 s2 adalah sinyal respons y1s1 "t" dipanggil sebagai x2s2

#sinyal respon S2
y1s2 = x1s2*x2s2 #sinyal respon y1 sistem S2 dipanggil sebagai y1s2
y2s2 = -x1s2*x2s2 #sinyal respon y2 sistem s2 dipanggil sebagai y2s2

#plot respon sistem s2
fig, ax = plt.subplots() #plot
plt.plot (x1s2, y1s2, label = "$y_1(t)$") #plot respon sistem S2 untuk y1
plt.plot (x1s2, y2s2, label = "$y_2(t)$") #plot respon sistem S2 untuk y2
plt.title ("2.Sinyal Respon S2 - MIMO") #memberikan judul pada sinyal respon sistem s2
plt.legend() #memberikan legenda pada plot sinyal respon sistem s2
plt.grid(True) #membuat grid pada plot

#3. Keluaran S3 adalah y1 = y1s2 untuk t>=0 dan y1 = y2s2 untuk t<0
#plot segmen rentang waktu
t2 = np.arange (-5, 0-dt, dt)#rentang waktu data terdefinisi x1s3 adalah dari -5 sampai 0-interval waktu sinyal dengan interval waktu 0,01 detik dipanggil sebagai t2
t1 = np.arange (0, 5, dt) #rentang waktu data terdefinisi x2s3 adalah dari 0 sampai 5 dengan interval waktu 0,01 detik dipanggil sebagai t1

#plot sinyal masukan S3
x1s3 = np.sin(t1) * t1 #sinyal masukan 1 pada s3 adalah sinyal respons 1 pada s2
x2s3 = -np.sin(t2) * t2 #sinyal masukan 2 pada s3 adalah sinyal respons 2 pada s2

#segmen sinyal respon S3
y1s3x1 = x1s3 #sinyal respon S3 untuk keluaran x1s3 dipanggil sebagai y1s3x1
y1s3x2 = x2s3 #sinyal respon S3 untuk keluaran x2s3 dipanggil sebagai y1s3x2

#plot respon sistem S3
fig, ax = plt.subplots() #plot
plt.plot (t1, y1s3x1, label = "$y_1(t) ; t>=0$") #plot respon sistem S3 untuk y1 rentang waktu t>=0
plt.plot (t2, y1s3x2, label = "$y_1(t) ; t<0$") #plot respon sistem S3 untuk y1 rentang waktu t<0
plt.title ("3.Sinyal Respon S3 - MISO") #memberikan judul pada sinyal respon sistem s3
plt.legend() #memberikan legenda pada plot sinyal respon sistem s3
plt.grid(True) #membuat grid pada plot
```

# SOAL 1 

```python
import numpy as np #memanggil modul numpy sebagai np
import pylab as plt #memanggil modul pylab sebagai plt

dt = 0.1 #interval waktu ditentukan 0.1 detik per data
t  = np.arange(-2, 2+dt, dt) #interval waktu adalah -2 sampai 2 dipanggil sebagai t
h  = np.sin(3*np.pi*t) * (np.heaviside(t, 1) - np.heaviside (t-5, 1)) #sinyal impuls respon dipanggil sebagai h
x  = np.cos(2*np.pi*t) * (np.heaviside(t, 1) - np.heaviside(t-5, 1)) #sinyal masukan dipanggil sebagai x

y  = np.convolve (x, h) * dt #hasil konvolusi h dan x dipanggil sebagai y
ty = np.arange (-4, 4+dt, dt) #rentang waktu grafik hasil konvolusi dari -4 sampai 4 dengan interval 0,01 detik dipanggil sebagai ty

fig, ax = plt.subplots() #plot
plt.plot (t, x, label='$x(t)$') #plot sinyal masukan
plt.plot (t, h, 'r-.', label='$h(t)$') #plot sinyal impuls respon
plt.title ("Plot x(t) dan h(t)") #memberikan judul pada hasil plot konvolusi
plt.legend()#memberikan legenda pada plot
plt.grid(True) #membuat grid pada plot

fig, ax = plt.subplots() #plot
plt.plot (ty, y) #plot hasil konvolusi
plt.title ("System Response") #memberikan judul pada hasil plot konvolusi
plt.grid(True) #membuat grid pada plot
plt.show()
```

```python
import numpy as np #memanggil modul numpy sebagai np
import pylab as plt #memanggil modul pylab sebagai plt

dt = 0.01 #interval waktu ditentukan 0.01 detik per data
t  = np.arange(-3.5, 3.5+dt, dt) #interval waktu adalah -3.5 sampai 3.5 dipanggil sebagai t
h  = 5 * np.exp(-t) * np.heaviside(t, 1)#sinyal impuls respon dipanggil sebagai h
x  = np.cos(np.pi*t) * (np.heaviside(t+2, 1) - np.heaviside(t-3, 1)) #sinyal masukan dipanggil sebagai x

y  = np.convolve (x, h) * dt #hasil konvolusi h dan x dipanggil sebagai y
ty = np.arange (-7, 7+dt, dt) #rentang waktu grafik hasil konvolusi dari -7 sampai 7 dengan interval 0,01 detik dipanggil sebagai ty

fig, ax = plt.subplots() #plot
plt.plot (t, x, label='$x(t)$') #plot sinyal masukan
plt.plot (t, h, 'r-.', label='$h(t)$') #plot sinyal impuls respon
plt.title ("Plot x(t) dan h(t)") #memberikan judul pada hasil plot konvolusi
plt.legend()#memberikan legenda pada plot
plt.grid(True) #membuat grid pada plot

fig, ax = plt.subplots() #plot
plt.plot (ty, y) #plot hasil konvolusi
plt.title ("Hasil Konvolusi Soal 2") #memberikan judul pada hasil plot konvolusi
plt.grid(True) #membuat grid pada plot
plt.show()
```

```python
import numpy as np #memanggil modul numpy sebagai np
import pylab as plt #memanggil modul pylab sebagai plt

#sinyal masukan
dt = 0.01 #interval waktu data 0,01 detik
tx1 = np.arange(0, 1+dt, dt) #rentang segmen waktu 1 sinyal masukan dipanggil sebagai tx1
tx2 = np.arange(1, 2+dt, dt) #rentang segmen waktu 2 sinyal masukan dipanggil sebagai tx2
tx  = np.concatenate((tx1, tx2)) #menggabungkan rentang waktu sinyal masukan dipanggil sebagai tx
x1  = tx1 #memanggil nilai segmen sinyal masukan 1 sebagai x1
x2  = -tx1+1 #memanggil nilai segmen sinyal masukan 2 sebagai x2
x   = np.concatenate((x1, x2)) #menggabungkan segmen sinyal masukan

#sinyal impuls respon
h1   = 1-tx1 #memanggil nilai segmen sinyal impuls respon sebagai h1
h2   = np.zeros_like(tx2) #memanggil nilai segmen sinyal impuls respon sebagai h2\
h    = np.concatenate((h1, h2))

#konvolusi
y  = np.convolve (x, h) * dt #hasil konvolusi h dan x dipanggil sebagai y
ty = np.arange (0, 4+3*dt, dt) #rentang waktu grafik hasil konvolusi dari diatur sedemikian rupa untuk akomodasi jumlah data konvolusi dipanggil sebagai ty

#plot sinyal masukan
fig, ax = plt.subplots() #plot pemisah
plt.plot (tx, x, label='$x(t)$') #plot sinyal masukan
plt.legend() #membuat legenda pada plot
plt.grid(True) #membuat grid pada plot

#plot sinyal masukan
fig, ax = plt.subplots() #plot pemisah
plt.plot (tx1, h1, label='$h(t)$') #plot sinyal impuls respon
plt.legend() #membuat legenda pada plot
plt.grid(True) #membuat grid pada plot

#plot konvolusi
fig, ax = plt.subplots() #plot pemisah
plt.plot (ty, y) #plot hasil konvolusi
plt.title ("Hasil Konvolusi") #memberikan judul pada hasil plot konvolusi
plt.grid(True) #membuat grid pada plot

plt.show()#menampilkan plot
```
