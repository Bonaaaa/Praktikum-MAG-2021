```python
import numpy as np #memanggil modul numpy sebagai np
import pylab as plt #memanggil modul pylab sebagai plt

#Interval waktu
dt = 0.01 #interval waktu sinyal adalah 0.01 detik per data
t  = np.arange(0,5+dt,dt) #rentang waktu data dari 0 sampai 5+interval waktu sinyal dengan interval waktu 0,01 detik

#sinyal masukan x(t)
x = t*np.cos(2*np.pi*t) #sinyal masukan didefinisikan sebagai "t cos (2*pi*t)"

#sinyal keluaran y(t)
y = 1.1*(t-2)*np.cos(2*np.pi*(t-2)) #sinyal keluaran didefinisikan sebagai 1,1*x(t-2)

#plot sinyal masukan
fig, ax = plt.subplots() #plot pemisah
plt.plot(t, x) #plot sinyal masukan
plt.title("Sinyal Masukan") #memberikan judul pada sinyal masukan
plt.grid(True) #membuat grid pada plot

#plot sinyal keluaran
fig, ax = plt.subplots() #plot pemisah
plt.plot(t, y) #plot sinyal keluaran
plt.title("Sinyal Keluaran") #memberikan judul pada sinyal keluaran
plt.grid(True) #membuat grid pada plot

plt.show()#menampilkan plot
```

```python
import numpy as np #memanggil modul numpy sebagai np
import pylab as plt #memanggil modul pylab sebagai plt

#Interval waktu
dt = 0.01 #interval waktu sinyal adalah 0.01 detik per data
t  = np.arange(-4,4+dt,dt) #rentang waktu data dari -4 sampai 4+interval waktu sinyal dengan interval waktu 0,01 detik

#3 sinyal masukan x(t)
x1 = np.heaviside (t,1)-np.heaviside (t-3,1) #sinyal masukan 1 didefinisikan sebagai "H(t)-H(t-3)"
x2 = t * np.sin (t) #sinyal masukan 2 didefinisikan sebagai "t * sin(t)"
x3 = t * np.cos (t) #sinyal masukan 3 didefinisikan sebagai "t * cos(t)"

#sinyal keluaran y(t)
y = x1 + (x2 * x3) #sinyal keluaran didefinisikan sebagai penjumlahan x1 dengan x2 * x3

#plot sinyal masukan
fig, ax = plt.subplots() #plot pemisah
plt.plot(t, x1, label = "$x_1(t)$") #plot sinyal masukan 1
plt.plot(t, x2, label = "$x_2(t)$") #plot sinyal masukan 2
plt.plot(t, x3, label = "$x_3(t)$") #plot sinyal masukan 3
plt.title("Sinyal Masukan") #memberikan judul pada sinyal-sinyal masukan
plt.legend() #memberikan legenda pada plot sinyal masukan
plt.grid(True) #membuat grid pada plot

#plot sinyal keluaran
fig, ax = plt.subplots() #plot pemisah
plt.plot(t, y) #plot sinyal keluaran
plt.title("Sinyal Keluaran") #memberikan judul pada sinyal keluaran
plt.grid(True) #membuat grid pada plot

plt.show()#menampilkan plot
```

```python
import numpy as np #memanggil modul numpy sebagai np
import pylab as plt #memanggil modul pylab sebagai plt

#Interval waktu
dt = 0.01 #interval waktu sinyal adalah 0.01 detik per data
t  = np.arange(-4,4+dt,dt) #rentang waktu data dari -4 sampai 4+interval waktu sinyal dengan interval waktu 0,01 detik

#sinyal masukan x(t)
x = 0.5 * np.heaviside (t,1) #sinyal masukan didefinisikan sebagai "0,5 * H(t)"

#2 sinyal keluaran y(t)
y1 = 0.5 * np.heaviside (t+2,1) #sinyal keluaran didefinisikan sebagai "0,5 * H(t+2)"
y2 = 0.5 * np.heaviside (t-2,1) #sinyal keluaran didefinisikan sebagai "0,5 * H(t-2)"

#plot sinyal masukan
fig, ax = plt.subplots() #plot pemisah
plt.plot(t, x) #plot sinyal masukan
plt.title("Sinyal Masukan") #memberikan judul pada sinyal-sinyal masukan
plt.grid(True) #membuat grid pada plot

#plot sinyal keluaran
fig, ax = plt.subplots() #plot pemisah
plt.plot(t, y1, label = "$y_1(t)$") #plot sinyal keluaran 1
plt.plot(t, y2, '--', label = "$y_2(t)$") #plot sinyal keluaran 2
plt.legend() #memberikan legenda pada plot sinyal sinyal keluaran
plt.title("Sinyal Keluaran") #memberikan judul pada sinyal keluaran
plt.grid(True) #membuat grid pada plot

plt.show()#menampilkan plot
```

```python
import numpy as np #memanggil modul numpy sebagai np
import pylab as plt #memanggil modul pylab sebagai plt

#Interval waktu
dt = 0.01 #interval waktu sinyal adalah 0.01 detik per data
t  = np.arange(-4,4+dt,dt) #rentang waktu data dari -4 sampai 4+interval waktu sinyal dengan interval waktu 0,01 detik

#2 sinyal masukan x(t)
x1 = np.heaviside (t,1) #sinyal masukan didefinisikan sebagai "H(t)"
x2 = 0.5 * np.heaviside (t-1,1) #sinyal masukan didefinisikan sebagai "0,5 * H(t-1)"

#2 sinyal keluaran y(t)
y1 = x1+x2 #sinyal keluaran didefinisikan sebagai penjumlahan sinyal masukan 1 dan 2
y2 = x1-x2 #sinyal masukan didefinisikan sebagai pengurangan sinyal masukan 1 dan 2

#plot sinyal masukan
fig, ax = plt.subplots() #plot pemisah
plt.plot(t, x1, label='$x_1(t)$') #plot sinyal masukan 1
plt.plot(t, x2, label='$x_2(t)$') #plot sinyal masukan 2
plt.title("Sinyal Masukan") #memberikan judul pada sinyal-sinyal masukan
plt.legend() #memberikan legenda pada plot sinyal sinyal masukan
plt.grid(True) #membuat grid pada plot

#plot sinyal keluaran
fig, ax = plt.subplots() #plot pemisah
plt.plot(t, y1, '-', label = "$y_1(t)$") #plot sinyal keluaran 1
plt.plot(t, y2, '--', label = "$y_2(t)$") #plot sinyal keluaran 2
plt.legend() #memberikan legenda pada plot sinyal sinyal keluaran
plt.title("Sinyal Keluaran") #memberikan judul pada sinyal keluaran
plt.grid(True) #membuat grid pada plot

plt.show()#menampilkan plot
```

```python
import numpy as np #memanggil modul numpy sebagai np
import pylab as plt #memanggil modul pylab sebagai plt

#segmen sinyal masukan, waktu diatur 2 kali rentang sinyal masukan didasari konsep konvolusi yang membalikkan dan menggeser sinyal
dt = 0.1 #interval waktu ditentukan 0.1 detik per data
tx1 = np.arange (-2, 0+dt, dt) #rentang waktu -2 sampai 0 dengan interval 0,1 dipanggil sebagai tx1
tx2 = np.arange (0, 2+dt, dt)#rentang waktu 0 sampai 2 dengan interval 0,1 dipanggil sebagai tx2
tx3 = np.arange (2, 4+dt, dt)#rentang waktu 2 sampai 4 dengan interval 0,1 dipanggil sebagai tx3
tx  = np.concatenate ((tx1, tx2, tx3)) #menggabungkan tx1, tx2, dan tx3
x1  = np.zeros_like(tx1) #nilai tx1 diatur bernilai nol sesuai soal dan dipanggil sebagai x1
x2  = np.zeros_like(tx2) + 1 #nilai tx2 diatur bernilai nol ditambah 1 sesuai soal dan dipanggil sebagai x2 
x3  = np.zeros_like(tx3) #nilai tx3 diatur bernilai nol sesuai soal dan dipanggil sebagai x3
x   = np.concatenate ((x1, x2, x3)) #gabungan x1,x2, dan x3 dipanggil sebagai x

#segmen sinyal masukan, waktu diatur 2 kali rentang sinyal masukan didasari konsep konvolusi yang membalikkan dan menggeser sinyal
th1 = np.arange(-2, 0+dt, dt) #rentang waktu -2 sampai 0 dengan interval 0,1 dipanggil sebagai th1
th2 = np.arange(0, 1+dt, dt) #rentang waktu 0 sampai 1 dengan interval 0,1 dipanggil sebagai th2
th3 = np.arange(1, 4+dt, dt) #rentang waktu 1 sampai 4 dengan interval 0,1 dipanggil sebagai th3
th  = np.concatenate((th1, th2, th3)) #menggabungkan th1, th2, th3
h1  = np.zeros_like(th1) #nilai h1 diatur bernilai nol sesuai soal dan dipanggil sebagai h1
h2  = 1-th2 #nilai h2 diatur 1-th2 sesuai soal dipanggil sebagai h2 
h3  = np.zeros_like(th3) #nilai h3 diatur bernilai nol sesuai soal ditambah 1 dan dipanggil sebagai x2 
h   = np.concatenate((h1, h2, h3)) #gabungan h1, h2, dan h3 dipanggil sebagai h

plt.plot (tx, x, label='$x(t)$') #plot sinyal masukan gabungan dari segmen yang didefinisikan sebelumnya
plt.plot (th, h, 'r-.', label='$h(t)$') #plot sinyal sistem gabungan dari segmen yang didefinisikan sebelumnya
plt.title("Plot x(t) dan h(t)") #memberikan judul pada plot
plt.legend()#memberikan legenda pada plot 
plt.grid(True) #membuat grid pada plot
plt.show()#menampilkan plot
```

```python
import numpy as np #memanggil modul numpy sebagai np
import pylab as plt #memanggil modul pylab sebagai plt

dt = 0.01 #interval waktu ditentukan 0.01 detik per data
t  = np.arange (0, 2+dt, dt) #rentang waktu dari 0 sampai 2 dan interval 0,01 detik dipanggil sebagai t
x  = np.ones((len(t))) #sinyal masukan dengan rentang waktu t diatur bernilai 1 dipanggil sebagai x

t1 = np.arange (0, 1+dt, dt) #rentang waktu 0 sampai 1 dan interval 0,01 detik dipanggil sebagai t1
t2 = np.arange (1, 2, dt) #rentang waktu 1 sampai 2 dan interval 0,01 detik dipanggil sebagai t2

h1 = 1-t1 #1-t1 dipanggil sebagai h1
h2 = np.zeros_like (t2) #t2 diatur bernilai nol dipanggil sebagai h2
h  = np.concatenate ((h1, h2)) #sistem dengan gabungan h1 dan h2 dipanggil sebagai h

y  = np.convolve(x, h) * dt #hasil konvolusi sinyal masukan dan sistem dipanggil sebagai y
ty = np.arange (0, 4+dt, dt) #rentang waktu grafik hasil konvolusi dari 0 sampai 4 dengan interval 0,01 detik dipanggil sebagai ty

plt.plot (ty,y) #plot grafik hasil konvolusi
plt.title("Hasil Konvolusi x(t) dan h(t)") #memberikan judul pada plot hasil konvolusi
plt.grid(True) #membuat grid pada plot
plt.show()#menampilkan plot hasil konvolusi
```
